```python
import pandas as pd
import requests
```

```python
oa = pd.read_csv(
    'data/Output_Area_to_Lower_Layer_Super_Output_Area_to_Middle_Layer_Super_Output_Area_to_Local_Authority_District_(December_2011)_Lookup_in_England_and_Wales(1).csv',
    index_col='OA11CD'
)
```

```python
ttwa = pd.read_csv(
    'data/Lower_Layer_Super_Output_Area_(2011)_to_Travel_to_Work_Area_(December_2011)_Lookup_in_the_United_Kingdom.csv',
    index_col='LSOA11CD'
).drop('ObjectId', axis=1).dropna()
```

```python
dz = pd.read_csv('https://www2.gov.scot/Resource/0046/00462937.csv', index_col='DataZone')
iz_names = pd.read_csv('https://opendata.arcgis.com/datasets/02f78aaf05df4f48a84cc5b10751ba3c_0.csv', index_col='IZ11CD') 
dz = dz.join(iz_names["IZ11NM"], on='InterZone')
dz
```

```python
dz2020 = pd.read_csv('http://statistics.gov.scot/downloads/file?id=5a9bf61e-7571-45e8-a307-7c1218d5f6b5%2FDatazone2011Lookup.csv', index_col='DataZone')
dz2020
```

```python
ni_lookup = pd.read_excel(
    'https://www.nisra.gov.uk/sites/nisra.gov.uk/files/publications/11DC_Lookup_1_1.xls',
    sheet_name='SOA2001_2_LGD2014',
    index_col='SOA2001'
)
ni_lookup
```

```python
lad_2020 = pd.read_csv("https://opendata.arcgis.com/datasets/fe6bcee87d95476abc84e194fe088abb_0.csv").drop('FID', axis=1)
lad_2020.index = lad_2020['LAD20CD']
lad_2020
```

```python
full = ttwa.join(lsoa.drop('LSOA11NM', axis=1)).reset_index()
countries = {
    "E": ("E92000001", "England"),
    "9": ("N92000002", "Northern Ireland"),
    "S": ("S92000003", "Scotland"),
    "W": ("W92000004", "Wales"),
}
full.loc[:, "CTRYCD"] = full["LSOA11CD"].apply(lambda x: countries.get(x[0])[0])
full.loc[:, "CTRYNM"] = full["LSOA11CD"].apply(lambda x: countries.get(x[0])[1])
full.loc[:, "MSOA11CD"] = full['LSOA11CD'].fillna(full["MSOA11CD"])
full.loc[:, "MSOA11NM"] = full['LSOA11NM'].fillna(full["MSOA11NM"])
full.loc[full["CTRYNM"]=="Scotland", ['MSOA11CD', "MSOA11NM"]] = (None, None)
full.loc[:, "MSOA11CD"] = full['MSOA11CD'].fillna(full.join(dz, on='LSOA11CD')['InterZone'])
full.loc[:, "MSOA11NM"] = full['MSOA11NM'].fillna(full.join(dz, on='LSOA11CD')['IZ11NM'])
full.loc[:, "LAD11CD"] = full['LAD11CD'].fillna(full.join(dz, on='LSOA11CD')['Council'])
full.loc[:, "LAD11CD"] = full['LAD11CD'].fillna(full.join(ni_lookup, on='LSOA11CD')["LGD2014"])
full.loc[:, "LAD11NM"] = full['LAD11NM'].fillna(full.join(ni_lookup, on='LSOA11CD')["LGD2014NAME"])
full = full.join(lad_2020, on='LAD11CD')
full
```

```python
la_successors = {}
for i, a in full.loc[
    full['LAD20CD'].isnull(),
    'LAD11CD',
].drop_duplicates().iteritems():
    la_successors[a] = set()
    r = requests.get('https://findthatpostcode.uk/areas/{}'.format(a))
    res = r.json()
    successors = list(set(res['data']['attributes']['successor']))
    for item in res['included']:
        if item.get('id') in successors:
            la_successors[a].add((item.get('id'), item['attributes']['name']))
    la_successors[a] = list(la_successors[a])
    if len(la_successors[a])==1:
        full.loc[
            full['LAD11CD']==a,
            ["LAD20CD", "LAD20NM"]
        ] = la_successors[a][0]
```

```python
lsoa_to_la2019 = pd.read_csv(
    "https://opendata.arcgis.com/datasets/4c6f3314565e43c5ac7885fd71347548_0.csv"
).set_index("LSOA11CD")[['UTLA19CD', 'UTLA19NM']]
full.loc[:, 'LAD20CD'] = full['LAD20CD'].fillna(full.join(lsoa_to_la2019, on='LSOA11CD')['UTLA19CD'])
full.loc[:, 'LAD20NM'] = full['LAD20NM'].fillna(full.join(lsoa_to_la2019, on='LSOA11CD')['UTLA19NM'])
full.loc[:, "LAD20CD"] = full['LAD20CD'].fillna(full.join(dz, on='LSOA11CD')['Council'])
```
